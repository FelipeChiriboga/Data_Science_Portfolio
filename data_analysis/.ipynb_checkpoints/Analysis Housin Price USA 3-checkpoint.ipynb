```python
import quandl
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
api_key = open('quandlapikey.txt', 'r').read()
```

```python
df = quandl.get("FMAC/HPI_AK")
df.head()
```

```python
fiddy_states = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states')
print(fiddy_states)
```

```python
fiddy_states = fiddy_states[0]
fiddy_states.head()
```

```python
fiddy_states = fiddy_states['Name &postal abbreviation[1]']['Name &postal abbreviation[1].1']
print(fiddy_states)
```

```python
for abbv in fiddy_states:
    print("FMAC/HPI_"+str(abbv))
```

```python
api_key = open('quandlapikey.txt','r').read()


main_df = pd.DataFrame()

for abbv in fiddy_states:
    query = "FMAC/HPI_"+str(abbv)
    df = quandl.get(query, authtoken=api_key)
    print(query)
    df[abbv] = (df[abbv] - df[abbv]) / df[abbv] * 100.0
        
    if main_df.empty:
        main_df = df
    else:
        main_df = main_df.join(df, lsuffix=abbv)
        
print(main_df.head())
```

```python
pickle_out = open('fiddy_states_3.pickle', 'wb')
pickle.dump(main_df, pickle_out)
pickle_out.close()
```

```python
pickle_in = open('fiddy_states_3.pickle')
HPI_data = pickle.load(pickle_in)
print(HPI_data)
```
